```python
import pyarrow.parquet as pq
# I use Colab here so I simply copy the data's path, I'll upload all the data this notebook need, if the path doesn't work, please change it manualy.

df1 = pq.read_table('./tf_flowers/embeddings/part-001.parquet').to_pandas()
df2 = pq.read_table('./tf_flowers/embeddings/part-002.parquet').to_pandas()
df3 = pq.read_table('./tf_flowers/embeddings/part-003.parquet').to_pandas()
df4 = pq.read_table('./tf_flowers/embeddings/part-004.parquet').to_pandas()
df5 = pq.read_table('./tf_flowers/embeddings/part-005.parquet').to_pandas()
df6 = pq.read_table('./tf_flowers/embeddings/part-006.parquet').to_pandas()
df7 = pq.read_table('./tf_flowers/embeddings/part-007.parquet').to_pandas()
df8 = pq.read_table('./tf_flowers/embeddings/part-008.parquet').to_pandas()
df9 = pq.read_table('./tf_flowers/embeddings/part-009.parquet').to_pandas()
df0 = pq.read_table('./tf_flowers/embeddings/part-000.parquet').to_pandas()
```

```python
import pandas as pd
frames = [df0, df1, df2, df3, df4, df5, df6, df7, df8, df9]
result = pd.concat(frames)
result = result.reset_index(drop=True)
```

```python
# I use Colab here so I simply copy the data's path, I'll upload all the data this notebook need, if the path doesn't work, please change it manualy.
with open('image_embeddings.txt', 'a') as f:
    for i in range(len(result)):
        f.write(str(result["image_name"][i]) + ' ')
        for j in result['embedding'][i]:
            f.write(str(j) + ' ')
        f.write('\n')
```

```python
# I use Colab here so I simply copy the data's path, I'll upload all the data this notebook need, if the path doesn't work, please change it manualy.
lines = open("./image_embeddings.txt").readlines()
sentences = []
```

```python
import re
for line in lines:
    line = re.sub(r'\[','',line)
    line = re.sub(r'\]','',line)
    sentences.append(line)
# I use Colab here so I simply copy the data's path, I'll upload all the data this notebook need, if the path doesn't work, please change it manualy.
with open('image_embeddings.txt', 'a') as f:
    for sentence in sentences:
        f.write(sentence)
```
