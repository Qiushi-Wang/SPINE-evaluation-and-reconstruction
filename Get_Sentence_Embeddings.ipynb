```python
import numpy as np
import pandas as pd
```

```python
import re
import string
import emoji
```

```python
# I use Colab here so I simply copy the data's path, I'll upload all the data this notebook need, if the path doesn't work, please change it manualy.
lines = open("../input/sentence/sentence.txt").readlines()
sentences = []
```

```python
for line in lines:
    #cleanTweet(line)
    #line = re.sub(r'#','',line)
    line = re.sub(r'@[A-Za-z0-9_]+','',line)
    line = re.sub(r'#','',line)
    line = re.sub(r'RT : ','',line)
    line = re.sub(r'\n','',line)
    # to remove emojis
    line = re.sub(emoji.get_emoji_regexp(), r"", line)
    line = re.sub(r'https?:\/\/[A-Za-z0-9\.\/]+','',line)
    line = re.sub(r"https?://\S+|www\.\S+","",line)
    line = re.sub(r"<.*?>","",line)
    line = str.lower(line)
    sentences.append(line)
```

```python
sentences[:5]
```

```python
data = []

for sentence in sentences:
    
    tokens = sentence.strip().split()
    data.append(tokens)
```

```python
data[:5]
```

```python
# I use Colab here so I simply copy the data's path, I'll upload all the data this notebook need, if the path doesn't work, please change it manualy.
lines = open("../input/original-embeddings/glove.6B.300d.txt").readlines()#[:limit] 
embeddings = []
words = []
for line in lines:
    tokens = line.strip().split()
    words.append(tokens[0])
    embeddings.append([float(i) for i in tokens[1:]])
```

```python
words[:5]
```

```python
#embeddings[:5]
```

```python
embeddings = np.array(embeddings)
```

```python
sentence_embeddings = []
for sentence in data:
    dup = 0
    sentence_embedding = np.zeros([1, 300])
    for word in sentence:
        if word in words:
            dup = dup + 1
            sentence_embedding = sentence_embedding + embeddings[words.index(word)]
    sentence_embedding = sentence_embedding / dup
    sentence_embeddings.append(sentence_embedding)
        
```

```python
head_sentences = []
for sentence in sentences:
    sentence = re.sub(r' ','_',sentence)
    head_sentences.append(sentence)
```

```python
head_sentences[:5]
```

```python
# I use Colab here so I simply copy the data's path, I'll upload all the data this notebook need, if the path doesn't work, please change it manualy.
assert len(sentence_embeddings) == len(sentences) #TODO print error statement
fw = open("sentence_embeddings.txt", 'w')
for i in range(len(sentence_embeddings)):
    fw.write(head_sentences[i] + " ")
    for j in sentence_embeddings[i][0]:
        fw.write(str(j) + " ")
    fw.write("\n")
fw.close()
```
